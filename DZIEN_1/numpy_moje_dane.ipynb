```python
m1 = np.array([[4,8,11],[67,101,3],[1,1,77]])
m2 = np.array([[499,9,3],[23,89,3],[2,4,11]])
wynik = (m1+m2)*m1

wynik
```

```python
print(f'wartosc minimum: {wynik.min()}')
print(f'wartosc maximum: {wynik.max()}')
print(f'wartosc średnia: {wynik.mean()}')
print(f'wartosc odchylenie standardowe: {wynik.std()}')
```
